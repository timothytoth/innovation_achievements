```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt



#sorting for readability
df['month_year'] = pd.to_datetime(df['month_year'])

# groupby month year and achievemetns(1,0) then take the mean of new repository
monthly_avg_new_repos = df.groupby(['month_year', 'has_achievements'])['New_Repository'].mean().unstack()

# Plotting 
plt.figure(figsize=(10, 6))
# Plot for specific pace colors woohoo so cool
plt.plot(monthly_avg_new_repos.index, monthly_avg_new_repos[0], marker='o', color='#00205B', label='Without Achievements')  # Deeper blue for users without achievements
plt.plot(monthly_avg_new_repos.index, monthly_avg_new_repos[1], marker='o', color='#FFD700', label='With Achievements')  # Vivid gold-yellow for users with achievements

#titles and labes
plt.title('Monthly Average of New Repositories')
plt.xlabel('Month-Year')
plt.ylabel('Average Number of New Repositories')
plt.xticks(rotation=45)
plt.legend(title='Achievement Status')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt


#  figure and subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6), sharey=True)

# Titles for each subplot
titles = ['New Repositories', 'Commits', 'Issue Events']

# Column names to plot
columns = ['New_Repository', 'Commits', 'IssuesEvent']

for i, col in enumerate(columns):
    # Group by month_year and has_achievements then calculate the mean
    monthly_avg = df.groupby(['month_year', 'has_achievements'])[col].mean().unstack()
    
    # pace colors
    axes[i].plot(monthly_avg.index, monthly_avg[0], marker='o', color='#00205B', label='Without Achievements')
    axes[i].plot(monthly_avg.index, monthly_avg[1], marker='o', color='#FFD700', label='With Achievements')
    
    # Setting the title for each subplot
    axes[i].set_title(titles[i])
    
  
    if i == 0:
        axes[i].legend(title='Achievement Status')
    
    # Formatting the x-axis dates
    axes[i].tick_params(labelrotation=45)
    axes[i].grid(True)

# Setting common labels
fig.text(0.5, 0.04, 'Month-Year', ha='center', va='center')
fig.text(0.06, 0.5, 'Average Count', ha='center', va='center', rotation='vertical')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Assuming df is your DataFrame

# Ensure 'month_year' is in datetime format
df['month_year'] = pd.to_datetime(df['month_year'])

# Creating the scatter plot
plt.figure(figsize=(12, 7))

# Plotting users without achievements
plt.scatter(df[df['has_achievements'] == 0]['month_year'], df[df['has_achievements'] == 0]['New_Repository'], color='#00205B', alpha=0.6, label='Without Achievements')

# Plotting users with achievements
plt.scatter(df[df['has_achievements'] == 1]['month_year'], df[df['has_achievements'] == 1]['New_Repository'], color='#FFD700', alpha=0.6, label='With Achievements')

# Formatting the date on the x-axis for better readability
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

# Adding plot title and labels
plt.title('Number of New Repositories Over Time by Achievement Status')
plt.xlabel('Time Period')
plt.ylabel('Number of New Repositories')

# Rotating date labels to avoid overlap
plt.xticks(rotation=45)

# Adding a legend
# Adding a legend with a fixed location
plt.legend(loc='upper right')

# Show plot
plt.show()
```
